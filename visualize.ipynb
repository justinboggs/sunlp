```python
import pandas as pd
from prompt_toolkit import prompt
import csv
import os
import seaborn as sns
import matplotlib.pyplot as plt
import investpy
from datetime import datetime
```

```python
site_kpis = pd.read_csv('data/site_kpis.csv')
site_kpis.head()
```

```python
fig, ax = plt.subplots()
ax.plot(site_kpis['Date'], site_kpis[['Price', 'Profit', 'Competitor_Price']])
ax.set_ylabel('Price')
ax.set_xlabel('Date')
ax.set_title('Price vs Competition')
plt.xticks(rotation=45)
ax.properties()['children'][1].set_color('black')
ax.properties()['children'][2].set_color('purple')
fig.set_size_inches(18.5, 10.5)
plt.grid()
ax.legend(['Price', 'Profit', 'Competitor Price'], loc='upper left')
ax.set_facecolor('xkcd:light grey')
```
